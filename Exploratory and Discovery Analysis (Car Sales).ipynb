# Exploratory Analysis

## Importing the relevant libraries

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import seaborn as sns
sns.set()
```

## Loading the raw data

```python
#Loading and creating a data frame for data set
raw_data = pd.read_csv('1.04. Real-life example.csv')
raw_data.head()
```

## Preprocessing

### Exploring the descriptive statistics of the variables

```python
#Summary  of 'Dirty' data frame
raw_data.describe(include='all')
```

### Determining the variables of interest

```python
#Dropping Column 'Model' because varying data types and low significance on the overall data set
data = raw_data.drop(['Model'],axis=1)
data.describe(include='all')
```

### Dealing with missing values

```python
#Checking for missing values in data frame and Summarizing null values for all features
data.isnull().sum()
```

```python
#Variable stands for 'Data no missing values', I am omitting the missing values from our distribution diagrams
data_no_mv = data.dropna(axis=0)
```

```python
data_no_mv.describe(include='all')
```

### Exploring the PDFs

```python
#Seaborn Visualization of feature 'Price'
sns.distplot(data_no_mv['Price'])
```

### Dealing with outliers

```python
#Outliers for Price in above diagram, therefore removing top 1%
q = data_no_mv['Price'].quantile(0.99)
data_1 = data_no_mv[data_no_mv['Price']<q]
data_1.describe(include='all')
```

```python
#Max Price is still Massive, but much closer than the Max Price from Cell 54
sns.distplot(data_1['Price'])
```

```python
#Before removing outliers for Feature 'Mileage'
sns.distplot(data_no_mv['Mileage'])
```

```python
#Removing top 1% for feature 'Mileage'
q = data_1['Mileage'].quantile(0.99)
data_2 = data_1[data_1['Mileage']<q]
```

```python
#Plotting
sns.distplot(data_2['Mileage'])
```

```python
#Plotting Feature 'Engine Volume'
sns.distplot(data_no_mv['EngineV'])
```

```python
# Engine sizes usually fall between 0.6 and 6.5, therefore we are removing outliers 
data_3 = data_2[data_2['EngineV']<6.5]
```

```python
#Plot updated 'EngineV'
sns.distplot(data_3['EngineV'])
```

```python
#Plot Feature 'Year'
sns.distplot(data_no_mv['Year'])
```

```python
#There does not seem to be many cars from 1970 so we are remvoing first Quartile
q = data_3['Year'].quantile(0.01)
data_4 = data_3[data_3['Year']>q]
```

```python
#Plotting Feature 'Year' after trimming
sns.distplot(data_4['Year'])
```

```python
#Here I am just resetting the ID's since we removed observations
data_cleaned = data_4.reset_index(drop=True)
```

```python
#Cleaned Data 
data_cleaned.describe(include='all')
```

## Checking the OLS assumptions

```python
#Checking for linearity
f, (ax1, ax2, ax3) = plt.subplots(1, 3, sharey=True, figsize =(15,3))
ax1.scatter(data_cleaned['Year'],data_cleaned['Price'])
ax1.set_title('Price and Year')
ax2.scatter(data_cleaned['EngineV'],data_cleaned['Price'])
ax2.set_title('Price and EngineV')
ax3.scatter(data_cleaned['Mileage'],data_cleaned['Price'])
ax3.set_title('Price and Mileage')


plt.show()
```

```python
#Price is not normally distributed
sns.distplot(data_cleaned['Price'])
```

### Relaxing the assumptions

```python
#Using Logrithmic Transformation to 'better fit' data
log_price = np.log(data_cleaned['Price'])
data_cleaned['log_price'] = log_price
data_cleaned
```

```python
#Plotting Calculations...seems much neater
f, (ax1, ax2, ax3) = plt.subplots(1, 3, sharey=True, figsize =(15,3))
ax1.scatter(data_cleaned['Year'],data_cleaned['log_price'])
ax1.set_title('Log Price and Year')
ax2.scatter(data_cleaned['EngineV'],data_cleaned['log_price'])
ax2.set_title('Log Price and EngineV')
ax3.scatter(data_cleaned['Mileage'],data_cleaned['log_price'])
ax3.set_title('Log Price and Mileage')


plt.show()
```

```python
#Price is no longer needed
data_cleaned = data_cleaned.drop(['Price'],axis=1)
```

### Multicollinearity

```python
#No Endogeniety will be observed later, Normality is assumed and central limit theorem, Homoscedasticity is solved with log transformation
data_cleaned.columns.values
```

```python
#SKLearn does not have anything for multicolinearity so I will use statsmodel

from statsmodels.stats.outliers_influence import variance_inflation_factor
variables = data_cleaned[['Mileage','Year','EngineV']]
vif = pd.DataFrame()
vif["VIF"] = [variance_inflation_factor(variables.values, i) for i in range(variables.shape[1])]
vif["features"] = variables.columns
```

```python
vif
```

```python
#Anything with a VIF lower than 10 we will keep, therefore Year is dropped
data_no_multicollinearity = data_cleaned.drop(['Year'],axis=1)
```

## Create dummy variables

```python
#Using pandas to automatically create dummy variables
data_with_dummies = pd.get_dummies(data_no_multicollinearity, drop_first=True)
```

```python
data_with_dummies.head()
```

### Rearrange a bit

```python
data_with_dummies.columns.values
```

```python
#Rearranging column names
cols = ['log_price', 'Mileage', 'EngineV', 'Brand_BMW',
       'Brand_Mercedes-Benz', 'Brand_Mitsubishi', 'Brand_Renault',
       'Brand_Toyota', 'Brand_Volkswagen', 'Body_hatch', 'Body_other',
       'Body_sedan', 'Body_vagon', 'Body_van', 'Engine Type_Gas',
       'Engine Type_Other', 'Engine Type_Petrol', 'Registration_yes']
```

```python
#Returning new Data Frame
data_preprocessed = data_with_dummies[cols]
data_preprocessed.head()
```

## Linear regression model

### Declare the inputs and the targets

```python
#Declaring inputs and targets
targets = data_preprocessed['log_price']
inputs = data_preprocessed.drop(['log_price'],axis=1)
```

### Scale the data

```python
#Standardize scale variables
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(inputs)
```

```python
inputs_scaled = scaler.transform(inputs)
```

### Train Test Split

```python
#Using 80/20 split
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(inputs_scaled, targets, test_size=0.2, random_state=365)
```

### Create the regression

```python
#Creating Regression for pre-processed data
reg = LinearRegression()
reg.fit(x_train,y_train)
```

```python
#Predictions
y_hat = reg.predict(x_train)
```

```python
plt.scatter(y_train, y_hat)
plt.xlabel('Targets (y_train)',size=18)
plt.ylabel('Predictions (y_hat)',size=18)
plt.xlim(6,13)
plt.ylim(6,13)
plt.show()
```

```python
#Using seaborn to observe distribution...seems to overestimate target judging from shift to the right
sns.distplot(y_train - y_hat)
plt.title("Residuals PDF", size=18)
```

```python
reg.score(x_train,y_train)
```

### Finding the weights and bias

```python
#Finding weights for Features to see which one has greatest/lowest impact
reg.intercept_
```

```python
reg.coef_
```

```python
#After printing, seems that Audi is benchmark, anything with positive is more expensive than Audi, and negative means less expensive than Audi
reg_summary = pd.DataFrame(inputs.columns.values, columns=['Features'])
reg_summary['Weights'] = reg.coef_
reg_summary
```

```python
data_cleaned['Brand'].unique()
```

## Testing

```python
#Testing our predictions
y_hat_test = reg.predict(x_test)
```

```python
#Alpha allows opacity to see concentration of points, almost a heat map of sorts
plt.scatter(y_test, y_hat_test, alpha=0.25)
plt.xlabel('Targets (y_test)',size=18)
plt.ylabel('Predictions (y_hat_test)',size=18)
plt.xlim(6,13)
plt.ylim(6,13)
plt.show()
```

```python
df_pf = pd.DataFrame(np.exp(y_hat_test), columns=['Prediction'])
df_pf.head()
```

```python
#Lots of missing values
df_pf['Target'] = np.exp(y_test)
df_pf
```

```python
#Forgetting original index helps with eliminating these missing values
y_test = y_test.reset_index(drop=True)
y_test.head()
```

```python
#Return Data Frame again
df_pf['Target'] = np.exp(y_test)
df_pf
```

```python
#Using this to see heart of algorithm
df_pf['Residual'] = df_pf['Target'] - df_pf['Prediction']
```

```python
#This finds differences in what was observed vs what was expected
df_pf['Difference%'] = np.absolute(df_pf['Residual']/df_pf['Target']*100)
df_pf
```

```python
#Stats summary of Prediction Data Frame
df_pf.describe()
```
